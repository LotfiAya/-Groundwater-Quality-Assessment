```python
import pandas as pd
from math import sqrt
df = pd.read_excel("WaterQuality.xlsx")
df
```

```python
si= [6.5 - 8.5, 30, 90, 3.00, 5.00, 1000.00, 20.00, 5.00, 0.5 ,50, 0.5, 5]
wi = [4, 1, 3, 4, 5, 4, 2, 2, 5, 5, 5, 5]
si1 = [6.5 - 8.5, 30, 3.00, 5.00, 50, 0.5, 5, 594]
wi1 = [4, 1 , 4, 5, 5, 5, 5, 5]
# Création des fonctions

## fonction Wi

def Wi(w, a):

    Wi = w[a] / sum(w)

    return Wi

## Calculons Qi :

def Qi(s, a, k):

    # Calculons ph pour le parametres pH

    if a == 0 :
        Qi = (df.loc[a, "pH"] - 8.5) / (6.5 - 8.5)

    # Les autres indices
    else :

        if a <= 11 :
            Qi = (df.loc[k, df.keys()[a+2]]*100)/ s[a]

    return Qi


## Calculons SI
def WAWQI():
    k = 0
    list_wawqi = []
    List_y = []

    # désigne les lignes
    for y in range(0, 18):
        wawqi = 0
        # x désigne les colonnes
        for x in range(0, 12):
            L = Wi(wi , x) * Qi(si, x, k)
            wawqi += L
        list_wawqi.append(wawqi)
        k += 1
        if wawqi <50:
            Y="Excellent  "
        elif wawqi>=50 and wawqi<=100:
            Y="Good "
        elif wawqi>=101 and wawqi<=200 :
            Y="poor "
        elif wawqi>=201 and wawqi<=300 :
            Y= "very poor "
        else:
            Y= "Unsuitable for drinking"
        List_y.append(Y)
    df['valeurs de WAWQI']= list_wawqi
    df['IND WAWQI']= List_y
    return df

WAWQI()

```

```python
def WGWQI():
    k = 0
    list_wgwqi = []
    List_y = []

    # désigne les lignes
    for y in range(0, 18):
        wgwqi = 1
        # x désigne les colonnes
        for x in range(0, 12):
            L = (Qi(si, x, k))**Wi(wi, x)
            wgwqi *= L
        list_wgwqi.append(wgwqi)
        k += 1
        if wgwqi <= 100 and wgwqi >= 90:
            Y="Excellent "
        elif wgwqi >= 70 and wgwqi <= 89:
            Y="Good "
        elif wgwqi >= 50 and wgwqi <= 69 :
            Y="Medium "
        elif wgwqi >= 25 and wgwqi <= 49 :
            Y= "Bad "
        elif wgwqi >= 0 and wgwqi <= 24:
            Y= "Very bad "
        List_y.append(Y)
    df['valeurs de WGWQI']= list_wgwqi
    df['IND WAWQI']= List_y
    return df
WGWQI()
```
