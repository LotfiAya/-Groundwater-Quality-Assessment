# Creation des class nessacaire pour le calcule 

**class WAWQI**

```python
class WAWQI :

  def __init__(self, Si, wi):
   
    self.si = Si
    self.wi = wi
    self.Calcul_WAWQI()

  def Calcul_Qi(self, b):

    if b == 0 :
      self.Qi = (df.loc[b, "pH"] - 8.5) / (6.5 - 8.5)

    else:
      if b <= 11 :
        self.Qi = (df.loc[self.IPL , df.keys()[b +2]]*100)/ self.si[b]
    
    return self.Qi
  
  def Calcul_wi(self, a):

    return self.wi[a]/sum(self.wi)
  
  def Calcul_WAWQI(self):
        self.IPL = 0
        self.total_SI = []

  #boucle sur les lignes

        for y in range(0, 18):
          SI = 0
          #boucle sur les colones
          for x in range(0, 12):
              chaymae = self.Calcul_wi(x) * self.Calcul_Qi(x)
              SI += chaymae
          self.total_SI.append(SI)
          self.IPL += 1
        df['valeurs de WAWQI']=self.total_SI
        print(df)
Si= [6.5 - 8.5, 30, 90, 3.00, 5.00, 1000.00, 20.00, 5.00, 0.5 ,50, 0.5, 5]
wi = [4, 1, 3, 4, 5, 4, 2, 2, 5, 5, 5, 5]
WAWQI(Si,wi)
  

```
